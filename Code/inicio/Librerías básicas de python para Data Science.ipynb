## Pandas

**Explicación Teórica:** Pandas es una biblioteca de Python para el análisis de datos. Proporciona estructuras de datos rápidas, flexibles y expresivas diseñadas para que trabajar con datos "relacionales" o "etiquetados" sea fácil e intuitivo. La librería se utiliza comúnmente para leer y escribir archivos de datos como CSV y Excel, manipular y transformar datos, realizar operaciones de limpieza de datos, y visualizar datos de manera efectiva.

Ejemplo:

El siguiente ejemplo muestra cómo leer un archivo CSV con Pandas, realizar algunas operaciones de limpieza y transformación de datos y finalmente visualizar los datos en un gráfico de barras utilizando la biblioteca Matplotlib.

```python
# Importar la biblioteca Pandas
import pandas as pd

# Crear un dataframe con algunos datos de ventas
data = {'region': ['Norte', 'Sur', 'Este', 'Oeste', 'Central'],
        'ventas': [10000, 8000, 12000, 9000, 11000],
        'ganancias': [5000, 4000, 6000, 4500, 5500]}
df = pd.DataFrame(data)

# Guardar el dataframe en un archivo CSV
df.to_csv('datos_ventas.csv', index=False)
```

En el código, primero importamos la biblioteca Pandas. Luego, creamos un diccionario data con algunos datos de ventas, y utilizamos el diccionario para crear un dataframe de Pandas llamado df.

Finalmente, utilizamos la función to_csv() para guardar el dataframe en un archivo CSV llamado datos_ventas.csv. El parámetro index=False se utiliza para no guardar el índice del dataframe en el archivo CSV.

```python
# Importar la biblioteca Pandas Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV y cargar los datos en un dataframe de Pandas
df = pd.read_csv('datos_ventas.csv')

# Imprimir las primeras 5 filas del dataframe
print(df.head())

# Eliminar las filas con valores faltantes
df.dropna(inplace=True)

# Agrupar los datos por región y calcular la suma de ventas para cada región
ventas_por_region = df.groupby('region')['ventas'].sum()
ganancias_por_region = df.groupby('region')['ganancias'].sum()

# Crear un gráfico de barras para visualizar las ventas por región
plt.bar(ventas_por_region.index, ventas_por_region.values)
plt.title('Ventas por región')
plt.xlabel('Región')
plt.ylabel('Ventas')
plt.show()

# Crear un gráfico de barras para visualizar las ventas por región
plt.bar(ventas_por_region.index, ganancias_por_region.values)
plt.title('Ganancias por región')
plt.xlabel('Región')
plt.ylabel('Ganancias')
plt.show()
```

El código primero importa la biblioteca Pandas y luego utiliza la función read_csv() para leer un archivo CSV llamado datos_ventas.csv y cargar los datos en un dataframe de Pandas llamado df. El código luego imprime las primeras 5 filas del dataframe utilizando la función head().

Después, el código utiliza la función dropna() para eliminar las filas con valores faltantes en el dataframe. Luego, utiliza la función groupby() para agrupar los datos por la columna region y calcular la suma de ventas para cada región.

Finalmente, el código utiliza la biblioteca Matplotlib para crear un gráfico de barras que muestra las ventas por región. El código utiliza las funciones bar(), title(), xlabel(), ylabel(), y show() para personalizar y mostrar el gráfico.

# Numpy

**Explicación Teórica:** Numpy es una biblioteca de Python para computación numérica. Proporciona una estructura de datos llamada array que es más eficiente que las listas estándar de Python para almacenar y manipular datos numéricos. Numpy también proporciona funciones matemáticas útiles para trabajar con matrices y vectores.

El siguiente ejemplo muestra cómo utilizar Numpy para crear una matriz aleatoria, realizar algunas operaciones matemáticas y visualizar los datos en una matriz de calor utilizando la biblioteca Matplotlib.

```python
# Importar la biblioteca Numpy y Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Crear una matriz aleatoria de 5x5
matriz = np.random.rand(5, 5)

# Imprimir la matriz
print(matriz)

# Calcular la media de cada fila de la matriz
medias = np.mean(matriz, axis=1)

# Imprimir las medias
print(medias)

# Visualizar la matriz en una matriz de calor
plt.imshow(matriz, cmap='hot')
plt.colorbar()
plt.show()
```

En el código, primero importamos las bibliotecas Numpy y Matplotlib. Luego, utilizamos la función random.rand() de Numpy para crear una matriz aleatoria de 5x5 llamada matriz.

Después, utilizamos la función mean() de Numpy para calcular la media de cada fila de la matriz. La opción axis=1 se utiliza para especificar que queremos calcular la media de cada fila en lugar de la media de cada columna.

Finalmente, utilizamos la función imshow() de Matplotlib para visualizar la matriz en una matriz de calor. La opción cmap='hot' se utiliza para especificar que queremos utilizar la paleta de colores "hot". La función colorbar() se utiliza para agregar una barra de color a la visualización.

# Matplotlib

Explicación Teórica: Matplotlib es una biblioteca de visualización de datos en Python. Se utiliza para crear gráficos de alta calidad en una variedad de formatos y estilos. Matplotlib proporciona una amplia gama de gráficos, desde simples gráficos de línea hasta gráficos de barras y diagramas de dispersión complejos.

Ejemplo:

El siguiente ejemplo muestra cómo utilizar Matplotlib para crear un gráfico de línea que muestra la evolución de las ventas a lo largo del tiempo utilizando los datos del archivo CSV generado en el ejemplo anterior.

```python
# Importar la biblioteca Pandas y Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV y cargar los datos en un dataframe de Pandas
df = pd.read_csv('datos_ventas.csv')

# Crear un gráfico de línea que muestra la evolución de las ventas a lo largo del tiempo
plt.plot(df['ventas'])
plt.title('Evolución de las ventas')
plt.xlabel('Tiempo')
plt.ylabel('Ventas')
plt.show()
```

El código primero importa la biblioteca Pandas y luego utiliza la función read_csv() para leer el archivo CSV generado en el ejemplo anterior y cargar los datos en un dataframe de Pandas llamado df.

Después, el código utiliza la biblioteca Matplotlib para crear un gráfico de línea que muestra la evolución de las ventas a lo largo del tiempo. El código utiliza la función plot() para crear el gráfico, y las funciones title(), xlabel(), ylabel() y show() para personalizar y mostrar el gráfico.

El gráfico muestra claramente la evolución de las ventas a lo largo del tiempo, lo que podría ayudar a identificar tendencias y patrones en los datos.

# Scikit-learn

Explicación Teórica: Scikit-learn es una biblioteca de Python para aprendizaje automático (machine learning). Proporciona herramientas para realizar tareas de aprendizaje supervisado y no supervisado, incluyendo clasificación, regresión, agrupamiento (clustering) y reducción de dimensionalidad. Scikit-learn también proporciona herramientas para la evaluación de modelos de aprendizaje automático y la selección de modelos.

Ejemplo:

El siguiente ejemplo muestra cómo utilizar Scikit-learn para entrenar un modelo de regresión lineal y hacer predicciones sobre nuevos datos.

```python
# Importar la biblioteca Pandas
import pandas as pd

# Crear un dataframe con algunos datos de ventas
data = {'region': ['Norte', 'Sur', 'Este', 'Oeste', 'Central'],
        'ventas': [10000, 8000, 12000, 9000, 11000],
        'ganancias': [5000, 4000, 6000, 4500, 5500]}
df = pd.DataFrame(data)

# Guardar el dataframe en un archivo CSV
df.to_csv('datos_ventas.csv', index=False)
```

```python
# Importar la biblioteca Pandas, Sklearn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Leer el archivo CSV y cargar los datos en un dataframe de Pandas
df = pd.read_csv('datos_ventas.csv')

# Dividir los datos en conjuntos de entrenamiento y prueba
X = df[['ganancias']]  # Característica (feature)
y = df['ventas']       # Etiqueta (label)

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, test_size=0.3, 
                                                                        random_state=0)
# Entrenar un modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_entrenamiento, y_entrenamiento)

# Hacer predicciones sobre los datos de prueba
y_pred = modelo.predict(X_prueba)

# Evaluar el modelo
print('Error cuadrático medio:', mean_squared_error(y_prueba, y_pred))
print('Coeficiente de determinación:', r2_score(y_prueba, y_pred))
```

# Seaborn

Explicación Teórica: Seaborn es una biblioteca de visualización de datos basada en Matplotlib. Proporciona una interfaz de alto nivel para crear gráficos estadísticos atractivos y informativos. Seaborn está diseñado para trabajar con datos complejos y de gran tamaño, y ofrece una amplia gama de gráficos, desde simples diagramas de dispersión hasta mapas de calor complejos.

Ejemplo:

El siguiente ejemplo muestra cómo utilizar Seaborn para visualizar los datos de ventas del archivo datos_ventas.csv creado anteriormente en un gráfico de barras agrupado.

```python
# Importar las bibliotecas Pandas y Seaborn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV y cargar los datos en un dataframe de Pandas
df = pd.read_csv('datos_ventas.csv')

# Crear un gráfico de barras agrupado con Seaborn
sns.barplot(x='region', y='ventas', data=df)
plt.title('Ventas por región')
plt.xlabel('Región')
plt.ylabel('Ventas')
plt.show()
```

En el código, primero importamos las bibliotecas Pandas y Seaborn. Luego, utilizamos la función read_csv() de Pandas para leer el archivo CSV datos_ventas.csv y cargar los datos en un dataframe de Pandas llamado df.

Después, utilizamos la función barplot() de Seaborn para crear un gráfico de barras agrupado que muestra las ventas por región. La función barplot() toma tres parámetros: x para especificar la columna que se utilizará para el eje x, y para especificar la columna que se utilizará para el eje y, y data para especificar el dataframe que se utilizará.

Finalmente, utilizamos las funciones title(), xlabel(), ylabel(), y show() para personalizar y mostrar el gráfico.
