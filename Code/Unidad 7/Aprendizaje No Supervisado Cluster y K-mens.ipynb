## Aprendizaje No Supervisado Cluster y K-mens

El aprendizaje no supervisado es una rama del aprendizaje automático que se utiliza cuando no se tienen datos etiquetados o clasificados previamente. En este tipo de aprendizaje, los algoritmos buscan patrones o estructuras subyacentes en los datos sin la ayuda de una variable de salida conocida.

El clustering, o agrupamiento, es una técnica de aprendizaje no supervisado que busca dividir un conjunto de datos en grupos, o clusters, basados en la similitud entre las instancias. El objetivo es que las instancias dentro de un mismo cluster sean similares entre sí, mientras que las instancias de clusters diferentes sean distintas.

K-means es uno de los algoritmos de clustering más populares. Este algoritmo busca dividir los datos en k clusters, donde k es un parámetro que se debe elegir previamente. K-means comienza seleccionando aleatoriamente k centroides, que representan el centro de cada uno de los clusters. Luego, el algoritmo asigna cada instancia al cluster cuyo centroide está más cerca. Después, los centroides se recalculan como la media de las instancias asignadas a cada cluster, y el proceso se repite hasta que no hay cambios en la asignación de instancias a clusters.

K-means es un algoritmo eficiente y fácil de implementar, pero tiene algunas limitaciones. Una de ellas es que el número de clusters k debe ser elegido previamente y de manera arbitraria, lo que puede ser difícil si no se tiene un conocimiento previo de los datos. Además, el algoritmo puede ser sensible a la inicialización aleatoria de los centroides, lo que puede llevar a resultados diferentes dependiendo de la inicialización.

En resumen, el clustering y el algoritmo k-means son herramientas útiles para encontrar patrones y estructuras subyacentes en conjuntos de datos no etiquetados, lo que puede ayudar a entender mejor los datos y a tomar decisiones informadas basadas en ellos. Sin embargo, es importante tener en cuenta las limitaciones de estos métodos y elegir cuidadosamente los parámetros y las inicializaciones para obtener resultados óptimos.

Para empezar con este ejercicio, primero debemos importar las bibliotecas necesarias. Usaremos numpy para crear algunos datos aleatorios y matplotlib para visualizarlos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Luego, crearemos algunos datos aleatorios para agruparlos. Usaremos numpy para generar 100 puntos aleatorios en un rango de -10 a 10.

```python
data = np.random.rand(100,2)*20-10
```

A continuación, visualizaremos estos datos para que podamos ver cómo están distribuidos.

```python
plt.scatter(data[:,0], data[:,1])
plt.show()
```

Podemos ver que estos datos no están agrupados de ninguna manera obvia. Ahora usaremos k-means para agruparlos.

Primero, importaremos la biblioteca para k-means:

```python
from sklearn.cluster import KMeans
```

Luego, ajustaremos el modelo k-means a los datos:

```python
model = KMeans(n_clusters=3)
model.fit(data)
```

Finalmente, visualizaremos los resultados:

```python
labels = model.predict(data)

plt.scatter(data[:,0], data[:,1], c=labels)
plt.show()
```

Podemos ver que k-means agrupó efectivamente los datos en tres grupos. ¡Este es solo un ejemplo básico de cómo se puede usar k-means para agrupar datos!
